```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/UnemploymentProject/CleanDatasets/bin_5_cleaned_future_rates.csv')
data.head()
data.info()
```

```python
for index, row in data.iterrows():
    year = row['Year']
    county = row['County']
    month = row['Month']

    for months_prev in [1, 2, 3]:
        if month - months_prev <= 0:
            prev_month = 12 + (month - months_prev)
            prev_year = year - 1
        else:
            prev_month = month - months_prev
            prev_year = year
        prev_rate = data[(data['Year'] == prev_year) &
                   (data['County'] == county) &
                   (data['Month'] == prev_month)]['Rate']

        if not prev_rate.empty:
            data.at[index, f'Rate_{months_prev}M_Prev'] = prev_rate.iloc[0]
        else:
            data.at[index, f'Rate_{months_prev}M_Prev'] = None

data.head(20)
```

```python
data = data.dropna()
data.head()
data.info()
```

```python
data.to_csv("/content/drive/MyDrive/UnemploymentProject/CleanDatasets/bin_5_cleaned_future_prev_rates.csv.csv", index=False)
```
