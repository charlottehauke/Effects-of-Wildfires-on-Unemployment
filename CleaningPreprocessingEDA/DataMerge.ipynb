```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
CPI = pd.read_csv('/content/drive/MyDrive/UnemploymentProject/CleanDatasets/CPI_clean.csv')
CPI = CPI[(CPI["Year"]>1989) & (CPI["Year"]<2017)]
CPI.info()
```

```python
em_nonfarm = pd.read_csv('/content/drive/MyDrive/UnemploymentProject/CleanDatasets/employees_on_nonfarm_payrolls_clean.csv')
em_nonfarm = em_nonfarm[(em_nonfarm["Year"]>1989) & (em_nonfarm["Year"]<2017)]
em_nonfarm.info()
```

```python
GDP = pd.read_csv('/content/drive/MyDrive/UnemploymentProject/CleanDatasets/CPI_clean.csv')
GDP = GDP[(GDP["Year"]>1989) & (GDP["Year"]<2017)]
GDP.info()
```

```python
ir = pd.read_csv('/content/drive/MyDrive/UnemploymentProject/CleanDatasets/interest_rate_clean.csv')
ir = ir.drop(columns='Year-Month')
ir.info()
```

```python
work_stoppages = pd.read_csv('/content/drive/MyDrive/UnemploymentProject/CleanDatasets/work_stoppages_clean.csv')
months = pd.DataFrame({'Month': range(1, 13)})
work_stoppages = work_stoppages.merge(months,how='cross')
work_stoppages.info()
```

```python
unemployment_fire = pd.read_csv('/content/drive/MyDrive/UnemploymentProject/CleanDatasets/unemployment_fire.csv')
unemployment_fire.info()
```

```python
merged_data = pd.merge(unemployment_fire, CPI, on=['Year', 'Month'], how='left')
merged_data = pd.merge(merged_data, em_nonfarm, on=['Year', 'Month'], how='left')
merged_data = pd.merge(merged_data, GDP, on=['Year', 'Month'], how='left')
merged_data = pd.merge(merged_data, ir, on=['Year', 'Month'], how='left')
merged_data = pd.merge(merged_data, work_stoppages, on=['Year', 'Month'], how='left')
```

```python
merged_data = merged_data.drop(columns=("CPI_x"))
merged_data = merged_data.rename(columns={'CPI_y':'CPI'})
merged_data.info()
merged_data.head(20)
```

```python
merged_data.reset_index(drop=True).to_csv("/content/drive/MyDrive/UnemploymentProject/CleanDatasets/merged_data.csv", index=False)
```
